## 1 Loading data

```python
import pandas as pd
```

```python
train_df = pd.read_csv("data/train.csv")
test_df = pd.read_csv("data/test.csv")
print(train_df.shape, test_df.shape)
```

A quick glance at the dataset:
- there are 1460 and 1459 data entries for train and test set, respectively
- there are missing values in both train and test set
- refer to data description, ``MSSubClass`` contains categorical information. However, its data type is stored as *int*

## 2 Data cleaning


2.1 Drop ``Id`` column since it is unnecessary for prediction

```python
train_df.drop("Id", axis=1, inplace=True)
test_df.drop("Id", axis=1, inplace=True)
```

2.2 Change ``MSSubClass``'s data type 

```python
train_df["MSSubClass"] = train_df["MSSubClass"].astype("object")
test_df["MSSubClass"] = test_df["MSSubClass"].astype("object")
```

```python
# collect the numerical and categorical features
cat_cols = [x for x in train_df.columns.to_list() if train_df[x].dtype == "object" and x != "SalePrice"]
num_cols = [x for x in train_df.columns.to_list() if x not in cat_cols and x != "SalePrice"]
print(len(num_cols), len(cat_cols))
```

2.3 Handling outliers  
Before imputing missing values, need to ensure that there aren't outliers (we don't want the missing values imputation is affected by outliers).  
Visualization using boxplot helps detecting outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axs = plt.subplots(5, 7, figsize=(30, 15))
axs = axs.ravel()
for i, col in enumerate(num_cols):
  sns.boxplot(x=col, data=train_df, ax=axs[i], color="blue")
```

Extreme outliers (= data points with large gap to other data points) are observed for ``LotFrontage``, ``LotArea``, ``BsmtFinSF1``, 
``TotalBsmtSF``, ``1stFlrSF``, ``GrLivArea``, ``MiscVal``. Remove the outliers from the train set:

```python
# drop data entries with extreme outliers
train_df = train_df.drop(train_df[train_df['LotFrontage'] > 200].index)
train_df = train_df.drop(train_df[train_df['LotArea'] > 100000].index)
train_df = train_df.drop(train_df[train_df['BsmtFinSF1'] > 4000].index)
train_df = train_df.drop(train_df[train_df['TotalBsmtSF'] > 4000].index)
train_df = train_df.drop(train_df[train_df['GrLivArea'] > 4000].index)
train_df = train_df.drop(train_df[train_df['MiscVal'] > 5000].index)

print(train_df.shape)
```

2.4. Handling missing values  
Now, check missing values at train and test data:

```python
# save target SalePrice
target = train_df["SalePrice"]

# combine train and test set
n_train = len(train_df)
all_df = pd.concat([train_df, test_df])
all_df.drop("SalePrice", axis=1, inplace=True)
print(all_df.shape)
```

```python
# calculate missing data
miss_data = {"Feature": [], "Total missing values": [], "Missing values (%)": []}
for c in all_df.columns.to_list():
  total_miss = all_df[c].isnull().sum()
  miss_percent = total_miss / len(all_df) * 100
  if miss_percent > 0:
    miss_data["Feature"].append(c)
    miss_data["Total missing values"].append(total_miss)
    miss_data["Missing values (%)"].append(miss_percent)
miss_df = pd.DataFrame(miss_data)
miss_df.sort_values("Missing values (%)", ascending=False)
```

Referring to data description, NA value in some columns, such as in ``PoolQC`` and ``MiscFeature``, is a value indicating non-existent of pool and misc feature in the house. So, handle missing values accordingly:

```python
# refer to data description, NA = None
to_none = ["PoolQC", "MiscFeature", "Alley", "Fence", "FireplaceQu", "GarageType", "GarageFinish", "GarageQual", "GarageCond", 
           "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2", "MasVnrType", "Utilities"]
for col in to_none:
  all_df[col].fillna("None", inplace=True)

# these features are related to features in to_none, so missing value means 0
to_zero = ["GarageArea", "GarageCars", "GarageYrBlt", "BsmtFinSF1", "BsmtFinSF2", "BsmtUnfSF", "TotalBsmtSF", "BsmtFullBath", "BsmtHalfBath", "MasVnrArea"]
for col in to_zero:
  all_df[col].fillna(0, inplace=True)

# NA can be assumed Type
all_df["Functional"].fillna("Typ", inplace=True)

# replace missing values in categorical column with the mode value
to_mode = ["MSZoning", "KitchenQual", "Electrical", "Exterior1st", "Exterior2nd", "SaleType"]
for col in to_mode:
  all_df[col].fillna(all_df[col].mode()[0], inplace=True)
```

```python
# LotFrontage
print(all_df["LotFrontage"].describe(), "\n\n", all_df["LotFrontage"].mode())
```

```python
# the mean and median of LotFrontage is similar, i.e. 68.995 and 68 -- replace with median
all_df["LotFrontage"].fillna(all_df["LotFrontage"].median(), inplace=True)
```

```python
# check if there is still missing values
for i in all_df.columns.to_list():
  if all_df[i].isnull().sum() > 0:
    print(i)
```

## 3 Exploratory data analysis and feature engineering

3.1 Organizing features  
Notice that there are several features with similar names. Let's check them:

```python
#TotBsmtSF = BsmtFinSF1 + BsmtFinSF2 + BsmtUnfSF
all(all_df["TotalBsmtSF"] == (all_df["BsmtFinSF1"] + all_df["BsmtFinSF2"] + all_df["BsmtUnfSF"]))
```

```python
# remove redundant Bsmt related columns
all_df.drop(labels=["BsmtFinSF1", "BsmtFinSF2", "BsmtUnfSF"], axis=1, inplace=True)
```

```python
# create new feature combining all bath-related features
all_df["TotalBath"] = all_df["FullBath"] + 0.5 * all_df["HalfBath"] + all_df["BsmtFullBath"] + 0.5 * all_df["BsmtHalfBath"]
# remove redundants
all_df.drop(["FullBath", "HalfBath", "BsmtFullBath", "BsmtHalfBath"], axis=1, inplace=True)

# create new feature combining all features related to porch
all_df["TotalPorch"] = all_df["OpenPorchSF"] + all_df["EnclosedPorch"] + all_df["3SsnPorch"] + all_df["ScreenPorch"]
# remove redundant porch columns
all_df.drop(["OpenPorchSF", "EnclosedPorch", "3SsnPorch", "ScreenPorch"], axis=1, inplace=True)
```

```python
# collect numerical columns
num_cols = [x for x in all_df.columns.to_list() if all_df[x].dtype != "object"]
print(len(num_cols))
```

With new features, let's visualize with histogram:

```python
# histogram
fig, axs = plt.subplots(4, 7, figsize=(30, 15))
axs = axs.ravel()
for i, col in enumerate(num_cols):
  sns.distplot(all_df[col], ax=axs[i], color="blue", norm_hist=True)
```

3.2 Fixing skewness  
Many features are skewed. Let's fix if the skewness is out of -0.5 ~ 0.5 range:

```python
from scipy.stats import skew, boxcox_normmax
from scipy.special import boxcox1p
import numpy as np
```

```python
# check skewness > abs(0.5)
skewed = {"Feature": [], "Skewness": []}
for col in num_cols:
  sk = skew(all_df[col])
  if abs(sk) > 0.5:
    skewed["Feature"].append(col)
    skewed["Skewness"].append(sk)
pd.DataFrame(skewed).head()
```

```python
print(len(skewed["Feature"]))
```

```python
# apply boxcox transformation with lambd = 0.15
for c in skewed["Feature"]:
  s = skew(all_df[c])
  if abs(s) > 0.5: 
    all_df[c] = boxcox1p(all_df[c], 0.15)
```

Also, need to check the target's distribution:

```python
# check skewness
print("Skewness: {}".format(skew(target)))
sns.distplot(target)
plt.show()
```

```python
# apply log transf. for target variable
target = np.log1p(target)
print("Skewness: {}".format(skew(target)))
sns.distplot(target)
plt.show()
```

3.3 Encoding categorical features  
Before proceeding to bivariate analysis, i.e. correlation check, let's encode the categorical features first:

```python
# encoding categorical features
all_df = pd.get_dummies(all_df, drop_first=True)
all_df.shape
```

3.4 Correlation check  
Correlation check is performed on the train set. Any change based on the correlation check will also be applied to test set

```python
# split back to train and test set
train_df = all_df[:n_train]
test_df = all_df[n_train:]

# concat target
corr_df = pd.concat([train_df[[*num_cols]], target], axis=1)
```

```python
# correlation matrix heatmap with target variable
corr_mat = corr_df.corr().sort_values("SalePrice", ascending=False)

fig, ax = plt.subplots(figsize=(20, 15))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr_mat, annot=True, cmap=cmap)
```

```python
# scatter plot of 2 features with high correlation to target
fig, axs = plt.subplots(1, 2, figsize=(12, 3))
axs = axs.ravel()
sns.regplot(x="OverallQual", y="SalePrice", data=corr_df, ax=axs[0])
sns.regplot(x="GrLivArea", y="SalePrice", data=corr_df, ax=axs[1])
```

Observations:  
- Strong correlation (> 0.7) to the target ``SalePrice`` is observed for ``OverallQual`` and ``GrLivArea``.   
- ``GarageCars``, ``GarageArea`` and ``GarageYrBlt`` are highly correlated among themselves. Although they are also strongly correlated to the target, we want to avoid multicollinearity -- select ``GarageCars``, and discard ``GarageArea`` and ``GarageYrBlt`` (because they have lower correlation to the target)
- ``GrLivArea`` is strongly correlated to ``TotRmsAbvGrd`` -- ``TotRmsAbvGrd`` will be discarded
- New feature ``TotalBath`` has a good correlation to the target, i.e. 0.67, whereas ``TotalPorch``'s correlation to target is only 0.3

```python
# discard GarageCars, GarageArea and GarageYrBuilt
train_df.drop(["GarageArea", "TotRmsAbvGrd", "GarageYrBlt"], axis=1, inplace=True)
test_df.drop(["GarageArea", "TotRmsAbvGrd", "GarageYrBlt"], axis=1, inplace=True)
print(train_df.shape, test_df.shape)
```

3.5 Finalize train and test set

```python
x_train = train_df
y_train = target
x_test = test_df
```

## 4 Modeling and evaluation  

```python
!pip install catboost
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.model_selection import KFold, cross_validate, RandomizedSearchCV
```

4.1 Cross validation on baseline models

```python
# baseline models
xgb = XGBRegressor(random_state=42)
lgb = LGBMRegressor(random_state=42)
cb = CatBoostRegressor(random_state=42, verbose=0)
```

```python
# cross validation
def crossVal(model):
  cv = KFold(10, shuffle=True, random_state=42)
  scores = cross_validate(model, x_train, y_train, scoring="neg_mean_squared_error", cv=cv, n_jobs=-1, return_train_score=True)
  train_score = "{:.5f}".format(np.sqrt(-scores["train_score"].mean()))
  val_score = "{:.5f}".format(np.sqrt(-scores["test_score"].mean()))
  return train_score, val_score

baseline_models = [xgb, lgb, cb]
for i in range(len(baseline_models)):
  print("----- Model {} -----".format(i + 1))
  train_score, val_score = crossVal(baseline_models[i])
  print("Train score: {}, validation score: {}".format(train_score, val_score))
```

- Each model produces comparable validation scores
- For each model, the training and validation MSE score are close --  all models don't suffer from either over- or underfitting

4.2 Hyperparameters tuning  
Although the baseline models are satisfactory, let's tune them to see if better models can be obtained:

```python
def hyperparameterTuning(model, params):
  cv = KFold(10, shuffle=True, random_state=42)
  search = RandomizedSearchCV(estimator=model, param_distributions=params, 
                              scoring="neg_mean_squared_error", n_jobs=-1, cv=cv, random_state=42)
  search.fit(x_train, y_train)
  best_rmse = "{:.5f}".format(np.sqrt(-search.best_score_))
  return search.best_params_, best_rmse
```

```python
# xgb model tuning
xgb_params = {
    "booster": ["gbtree", "dart"],
    "learning_rate": [0.05, 0.1, 0.5],
    "max_depth": [3, 5, 7],
    "min_child_weight": [1e-2, 0.1, 1],
    "subsample": [0.5, 0.7, 1],
    "colsample_bytree": [0.5, 0.7, 1],
    "colsample_bynode": [0.5, 0.7, 1],
    "colsample_bylevel": [0.5, 0.7, 1],
    "n_estimators": [500, 1000, 2000, 3000], 
    "reg_alpha": [1e-2, 0.1, 1],
    "reg_lambda": [1e-2, 0.1, 1], 
    "objective": ["reg:squarederror", "reg:linear"], 
    "importance_type": ["weight", "gain", "cover"]
}

best_par, best_score = hyperparameterTuning(xgb, xgb_params)
print(best_score)
print(best_par)
```

```python
# lgb model tuning
lgb_params = {
    "boosting_type": ["gbdt", "dart", "goss", "rf"],
    "num_leaves": [15, 30, 45],
    "learning_rate": [0.05, 0.1, 0.5],
    "max_depth": [-1, 3, 5, 7],
    "min_child_weight": [1e-3, 1e-2, 0.1],
    "min_child_samples": [10, 20, 30],
    "subsample": [0.5, 0.7, 1],
    "colsample_bytree": [0.5, 0.7, 1],
    "n_estimators": [500, 1000, 2000, 3000],
    "reg_alpha": [1e-2, 0.1, 1],
    "reg_lambda": [1e-2, 0.1, 1], 
    "importance_type": ["split", "gain"]
}

best_par, best_score = hyperparameterTuning(lgb, lgb_params)
print(best_score)
print(best_par)
```

```python
# catboost model tuning
cb_params = {"learning_rate": [0.05, 0.1, 0.5],
             "reg_lambda": [1e-2, 0.1, 1], 
             "max_depth": [3, 5, 7],
             "n_estimators": [500, 1000, 2000, 3000],
             "subsample": [0.5, 0.7, 1], 
             "colsample_bylevel": [0.5, 0.7, 1],
             "min_child_samples": [10, 20, 30]
}

best_par, best_score = hyperparameterTuning(cb, cb_params)
print(best_score)
print(best_par)
```

The cross validation result of all three models are close

4.3 Training the best models

```python
# define best models
best_xgb = XGBRegressor(subsample=0.7, reg_lambda=1, reg_alpha=0.1, objective="reg:linear", 
                        n_estimators=500, min_child_weight=1, max_depth=3, learning_rate=0.05, importance_type="gain",
                        colsample_bytree=1, colsample_bynode=0.5, colsample_bylevel=0.7, booster="dart")

best_lgb = LGBMRegressor(subsample=0.5, reg_lambda=0.01, reg_alpha=0.01, num_leaves=30, 
                         n_estimators=1000, min_child_weight=0.01, min_child_samples=30,
                         max_depth=3, learning_rate=0.05, importance_type="gain", colsample_bytree=0.7,
                         boosting_type="gbdt")

best_cb = CatBoostRegressor(subsample=0.5, reg_lambda=1, n_estimators=1000, min_child_samples=10, max_depth=5, learning_rate=0.05,
                            colsample_bylevel=0.7, verbose=0)

# train
best_xgb.fit(x_train, y_train)
best_lgb.fit(x_train, y_train)
best_cb.fit(x_train, y_train)
```

4.4 Test predicting train set  
Just to see how the models predict the train set:

```python
from sklearn.metrics import mean_squared_error

def rmse(y_pred):
  return np.sqrt(mean_squared_error(y_train, y_pred))
```

```python
y_pred_xgb = best_xgb.predict(x_train)
y_pred_lgb = best_lgb.predict(x_train)
y_pred_cb = best_cb.predict(x_train)

rmse_xgb = rmse(y_pred_xgb)
rmse_lgb = rmse(y_pred_lgb)
rmse_cb = rmse(y_pred_cb)

print("XGB Score: {}".format(rmse_xgb))
print("LGB Score: {}".format(rmse_lgb))
print("CB Score: {}".format(rmse_cb))
```

- The performance on the train set is satisfactory
- CatBoost model achieves the lowest RMSE score when predicting train set  

Looking back at the cross validation result after hyperparameters tuning, all three models are comparable. So, another strategy that is worth to try is averaging the predictions (averaging predictions is an ensemble method as well):

```python
# avg
y_pred_avg = (y_pred_xgb + y_pred_lgb + y_pred_cb) / 3
print("Average prediction score: {}".format(rmse(y_pred_avg)))
```

## 5 Prediction

```python
y_pred_xgb = best_xgb.predict(x_test)
y_pred_lgb = best_lgb.predict(x_test)
y_pred_cb = best_cb.predict(x_test)
y_pred_avg = (y_pred_xgb + y_pred_lgb + y_pred_cb) / 3

y_pred_cb = np.exp(y_pred_cb)
y_pred_avg = np.exp(y_pred_avg)
```
